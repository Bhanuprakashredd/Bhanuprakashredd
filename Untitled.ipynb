```python
# Install required libraries (if not already installed)
!pip install pytesseract opencv-python
!sudo apt install tesseract-ocr
!sudo apt install tesseract-ocr-san # Install Sanskrit language pack

# For Google Colab, you need to mount your Google Drive
from google.colab import drive
drive.mount('/content/drive')

import os
import cv2
import pytesseract
import pandas as pd
import re

# Define the path to your dataset folder in Google Drive
# Replace with the appropriate path if needed.
# Based on the provided link, adjust the path to where your images are stored.
dataset_folder = '/content/drive/MyDrive/Colab Notebooks/Dataset'  # Update with your folder path

# Output CSV file path
output_csv = '/content/drive/MyDrive/sanskrit_ocrtext.csv'

# Configure Tesseract to use Sanskrit language; ensure that 'tesseract-ocr-san' is installed.
# The language code 'san' corresponds to Sanskrit.
custom_config = r'--oem 3 --psm 6 -l san'

# Prepare a list to store the results
ocr_results = []

# Loop through each file in the dataset folder
for filename in os.listdir(dataset_folder):
    if filename.lower().endswith(('.jpg', '.jpeg', '.png')):
        image_path = os.path.join(dataset_folder, filename)

        # Read the image using OpenCV
        image = cv2.imread(image_path)

        # Check if the image was loaded properly
        if image is None:
            print(f"Unable to load image: {filename}")
            continue

        # Apply OCR to extract text from the image
        extracted_text = pytesseract.image_to_string(image, config=custom_config)

        # Use regex to keep only Devanagari characters (Unicode range U+0900 to U+097F) and whitespace.
        # This removes numbers and other characters.
        sanskrit_text = re.sub(r'[^\u0900-\u097F\s]', '', extracted_text)

        # Remove extra whitespace
        sanskrit_text = re.sub(r'\s+', ' ', sanskrit_text).strip()

        # Append the result (filename and cleaned Sanskrit text) to the list
        ocr_results.append({'Filename': filename, 'Extracted_Sanskrit_Text': sanskrit_text})

# Convert the results into a pandas DataFrame
df = pd.DataFrame(ocr_results)

# Save the DataFrame to CSV
df.to_csv(output_csv, index=False, encoding='utf-8')

print(f"✅ OCR processing complete. Extracted Sanskrit text saved to: {output_csv}")
```

```python
import pandas as pd
import re
import unicodedata
import random  # 🔧 Needed for masking
from google.colab import drive

# ✅ Mount Google Drive
drive.mount('/content/drive')

# ✅ File paths
ocr_csv_path = '/content/drive/MyDrive/sanskrit_ocrtext.csv'
cleaned_csv_path = '/content/drive/MyDrive/cleaned_sanskrit_text.csv'
paired_training_data_path = '/content/drive/MyDrive/sanskrit_training_pairs.csv'

# ✅ Load OCR CSV
df = pd.read_csv(ocr_csv_path)

# ✅ Cleaning and normalization function
def clean_and_normalize(text):
    if pd.isna(text):
        return ''

    # Common OCR misread corrections
    ocr_corrections = {
        'l': '।',      # Small 'l' mistaken for danda
        '|': '।',      # Pipe mistaken for danda
        "'": 'ऽ',      # Apostrophe as avagraha
        '`': '',       # Remove stray grave accents
    }
    for wrong, right in ocr_corrections.items():
        text = text.replace(wrong, right)

    # Normalize Unicode
    text = unicodedata.normalize('NFC', text)

    # Remove zero-width characters (invisible junk)
    text = re.sub(r'[\u200B-\u200D\uFEFF]', '', text)

    # Keep only Devanagari chars, space, danda
    text = re.sub(r'[^\u0900-\u097F\s।]', '', text)

    # Normalize spacing & punctuation
    text = re.sub(r'[।]{2,}', '।', text)
    text = re.sub(r'\s+।', '।', text)
    text = re.sub(r'।\s*', '। ', text)
    text = re.sub(r'\s+', ' ', text)
    text = text.strip()

    return text

# ✅ Apply cleaning
df['Cleaned_Text'] = df['Extracted_Sanskrit_Text'].apply(clean_and_normalize)

# ✅ Save cleaned text
df.to_csv(cleaned_csv_path, index=False, encoding='utf-8')
print(f"✅ Cleaned Sanskrit text saved to: {cleaned_csv_path}")

# ✅ Masking function
def mask_random_word(text):
    words = text.split()
    if len(words) < 3:
        return text, text  # Skip masking for very short texts
    index_to_mask = random.randint(1, len(words) - 2)
    masked = words[:index_to_mask] + ['<mask>'] + words[index_to_mask + 1:]
    return ' '.join(masked), text

# ✅ Create masked pairs
masked_data = [mask_random_word(t) for t in df['Cleaned_Text']]
paired_df = pd.DataFrame(masked_data, columns=['Masked_Input', 'Original_Text'])

# ✅ Save masked training pairs
paired_df.to_csv(paired_training_data_path, index=False, encoding='utf-8')
print(f"✅ Masked training pairs saved to: {paired_training_data_path}")

# ✅ Preview
print("\n🔍 Sample masked data:\n")
print(paired_df.head(10))
```

```python
from transformers import AutoModelForSeq2SeqLM, Seq2SeqTrainer, Seq2SeqTrainingArguments, DataCollatorForSeq2Seq
from datasets import Dataset
import torch

# ✅ Load the dataset
data = pd.read_csv('/content/drive/MyDrive/sanskrit_training_pairs.csv')
dataset = Dataset.from_pandas(data)

# ✅ Tokenize
def preprocess_function(examples):
    model_inputs = tokenizer(
        examples["Masked_Input"],
        padding="max_length",
        max_length=128,
        truncation=True
    )
    labels = tokenizer(
        examples["Original_Text"],
        padding="max_length",
        max_length=128,
        truncation=True
    )

    model_inputs["labels"] = labels["input_ids"]
    return model_inputs

tokenized_dataset = dataset.map(preprocess_function, batched=True)

# ✅ Load model
model = AutoModelForSeq2SeqLM.from_pretrained("google/byt5-small")

# ✅ Training configuration
# Replacing 'evaluation_strategy' with 'eval_strategy'
training_args = Seq2SeqTrainingArguments(
    output_dir="/content/byt5-sanskrit-checkpoints",
    eval_strategy="epoch",  # Changed to 'eval_strategy'
    learning_rate=5e-5,
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    weight_decay=0.01,
    save_total_limit=2,
    num_train_epochs=5,
    fp16=torch.cuda.is_available(),
    # Remove or comment out 'evaluation_strategy' if present
    # evaluation_strategy="epoch",
)


# ✅ Data collator
data_collator = DataCollatorForSeq2Seq(tokenizer, model=model)

# ✅ Trainer setup
trainer = Seq2SeqTrainer(
    model=model,
    args=training_args,
    train_dataset=tokenized_dataset,
    eval_dataset=tokenized_dataset.select(range(200)),  # Just for quick evaluation
    tokenizer=tokenizer,
    data_collator=data_collator,
)

# ✅ Train the model
trainer.train()
```
